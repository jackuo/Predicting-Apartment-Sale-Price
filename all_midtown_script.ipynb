```python
from bs4 import BeautifulSoup
import requests
import urllib.request

from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
import time

import os
chromedriver = "/Applications/chromedriver" # path to the chromedriver executable
os.environ["webdriver.chrome.driver"] = chromedriver

import pickle
import random
import sys
```

```python
# Using City Realty Website
url = 'https://www.cityrealty.com/nyc/closed-sales/search-results#?page=1'
```

```python
### Try with Selenium
driver = webdriver.Chrome(chromedriver)
driver.get(url)
```

```python
### Go through pages and pull out URLs

url_list = []

for page_val in range(1, 460): # hard coded range based on total number of closing sales, and assume 50 entries per page
    
    soup = BeautifulSoup(driver.page_source, 'html.parser') # pull out HTML

    for row in soup.find_all('a', {"ng-if" : "!!listing.url_listing"}): #Iterate through every entry containing URL
        url = row.get('href') # grab URL
        url_list.append(url) # append URL to list
    
    # go to next page
    next_page = str(page_val + 1) # identify next page for button
    x_path_string = '//a[text()={}]'.format(next_page) # create x_path
    next_page_button = driver.find_element_by_xpath(x_path_string) # find element
    
    # scroll to page
    driver.execute_script("return arguments[0].scrollIntoView();", next_page_button)
    driver.execute_script("window.scrollBy(0, -150);")
    
    time.sleep(1) # give time to move to right location on page
    next_page_button.click() # click next page button, and wait
    
    # Print for sanity check
    print("Page: ", page_val)
    print("Size of soup: ", str(sys.getsizeof(soup)))
    
    time.sleep(3+4*random.random())
```
