```python
import time
import random

ua = UserAgent()
user_agent = {'User-agent': ua.random}

count = 1

for apt in close_list:
    
    try:
        response = requests.get(apt, headers = user_agent)
        response_list = [response.text, response.url]
        filename = '/Users/JacKuo14/Documents/Metis/course_work/midtown_data_dump/apt_{}.pkl'.format(count)
        #filename = "/Users/JacKuo14/Documents/Metis/course_work/city_realty_data/all_midtown/apt_{}.pkl".format(count)
        
        with open(filename, 'wb') as picklefile:
            pickle.dump(response_list, picklefile)
        
        print("Good" + " " + str(count))
        count = count + 1
        
        time.sleep(2 + 2*random.random())
            
    except requests.exceptions.RequestException as e:
        print(apt)
        continue
```
